## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
import seaborn as sns
sns.boxplot(df['ConvertedComp'])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1
print(f'inter quartile range is {q3 - q1}')
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = q1 - 1.5 * (iqr)
upper_bound = q3 + 1.5 * (iqr)

print(f'lower bound: {lower_bound}\nupper bound: {upper_bound}')
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df_outlier = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
df_outlier.shape
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_clean = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
df_clean.head()
```

```python
df_clean['ConvertedComp'].median()
df_clean['ConvertedComp'].mean()
```
